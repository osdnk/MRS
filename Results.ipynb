## Graphs
Let us notice that pyhon gives us *hardly-inifite* precision and seems to be nice for our case
Error is evaluating in fortan where u(x) = x in kind=16 precision

### kind = 4

```python
errors = []
for i in range(2,1000):
    cmd = f"./FDM4 {i}"
    r = !{cmd}
    errors.append(float(r[0].split()[0]))
```

```python
axes = plt.gca()
axes.set_ylim([0,0.0001])
plt.scatter(range(2,1000), errors)
plt.show()
```
